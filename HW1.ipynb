```python
import pandas as pd

# Load the dataset
df = pd.read_csv("Baltimore911.csv")

# Data types and summary statistics
for column in df.columns:
    # Data type
    data_type = df[column].dtype
    
    # Determine if the column is numeric or discrete
    if pd.api.types.is_numeric_dtype(df[column]):
        attribute_type = "Numeric"
        min_value = df[column].min()
        max_value = df[column].max()
        mean_value = df[column].mean()
        median_value = df[column].median()
        std_dev = df[column].std()
    else:
        attribute_type = "Discrete"
        unique_values = df[column].nunique()
        top_3_values = df[column].value_counts().head(3)
        
    # Missing values
    missing_percentage = (df[column].isnull().sum() / len(df)) * 100
    
    # Print column analysis
    print(f"Column: {column}")
    print(f"Data Type: {data_type}")
    print(f"Attribute Type: {attribute_type}")
    if attribute_type == "Numeric":
        print(f"Min: {min_value}")
        print(f"Max: {max_value}")
        print(f"Mean: {mean_value}")
        print(f"Median: {median_value}")
        print(f"Standard Deviation: {std_dev}")
    else:
        print(f"Total Unique Values: {unique_values}")
        print("Top 3 Attribute Values with Largest Count:")
        print(top_3_values)
    print(f"Missingness: {missing_percentage:.2f}%\n")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("Baltimore911.csv")

# Convert the CrimeDate column to datetime format
df['CrimeDate'] = pd.to_datetime(df['CrimeDate'])

df_filtered = df[df['CrimeDate'].dt.year >= 2014]

# Extract year, month, and day of the week from the CrimeDate column
df_filtered['Year'] = df_filtered['CrimeDate'].dt.year
df['Month'] = df['CrimeDate'].dt.month
df['DayOfWeek'] = df['CrimeDate'].dt.dayofweek

# Group by year, month, and day of the week and count the number of incidents
yearly_incidents = df_filtered.groupby('Year').size()
monthly_incidents = df.groupby('Month').size()
day_of_week_incidents = df.groupby('DayOfWeek').size()

# Plotting the data
plt.figure(figsize=(15, 5))

# Year-to-year crime incidence
plt.subplot(1, 3, 1)
yearly_incidents.plot(kind='bar', color='skyblue')
plt.title('Year-to-Year Crime Incidence')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)

# Monthly crime incidence
plt.subplot(1, 3, 2)
monthly_incidents.plot(kind='bar', color='orange')
plt.title('Monthly Crime Incidence')
plt.xlabel('Month')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=0)

# Day of the week crime incidence
day_names = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
plt.subplot(1, 3, 3)
day_of_week_incidents.index = day_names
day_of_week_incidents.plot(kind='bar', color='green')
plt.title('Day of the Week Crime Incidence')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("Baltimore911.csv")

# Convert the CrimeDate column to datetime format
df['CrimeDate'] = pd.to_datetime(df['CrimeDate'])

# Extract year, month, and day of the week from the CrimeDate column
df['Year'] = df['CrimeDate'].dt.year
df['Month'] = df['CrimeDate'].dt.strftime('%Y-%m')
df['DayOfWeek'] = df['CrimeDate'].dt.strftime('%A')

# Define the order of days for sorting
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Filter data for the years 2014 to 2020
df_filtered = df[(df['Year'] >= 2014) & (df['Year'] <= 2020)]

# Group by Year and Crime Type
yearly_crime_counts = df_filtered.groupby(['Year', 'Description']).size().reset_index(name='Count')

# Group by Month and Crime Type
monthly_crime_counts = df_filtered.groupby(['Month', 'Description']).size().reset_index(name='Count')

# Group by Day of the Week and Crime Type
day_of_week_crime_counts = df_filtered.groupby(['DayOfWeek', 'Description']).size().reindex(days_order, level=0).reset_index(name='Count')

# Plotting the data
plt.figure(figsize=(15, 12))

# Crime type distributions by year (starting from 2014)
plt.subplot(3, 1, 1)
for crime_type, data in yearly_crime_counts.groupby('Description'):
    plt.plot(data['Year'], data['Count'], marker='o', label=crime_type, linewidth=2)
plt.title('Crime Type Distributions by Year (2014 to 2020)')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)
plt.legend(title='Crime Type')

# Crime type distributions by month
plt.subplot(3, 1, 2)
for crime_type, data in monthly_crime_counts.groupby('Description'):
    plt.plot(data['Month'], data['Count'], marker='o', label=crime_type, linewidth=2)
plt.title('Crime Type Distributions by Month (2014 to 2020)')
plt.xlabel('Month')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)
plt.legend(title='Crime Type')

# Crime type distributions by day of the week
plt.subplot(3, 1, 3)
for crime_type, data in day_of_week_crime_counts.groupby('Description'):
    plt.plot(data['DayOfWeek'], data['Count'], marker='o', label=crime_type, linewidth=2)
plt.title('Crime Type Distributions by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)
plt.legend(title='Crime Type')

plt.tight_layout()
plt.show()
```
